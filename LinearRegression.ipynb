```python
import pandas
from pandas import DataFrame
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
```

```python
data = pandas.read_csv("cost_revenue_clean.csv")
```

```python
data
```

```python
x = DataFrame(data, columns=["production_budget_usd"])
y = DataFrame(data, columns=["worldwide_gross_usd"])
```

```python
plt.figure(figsize = (10, 6))
plt.scatter(x,y, alpha=0.3)
plt.title("Film production cost and global revenue")
plt.xlabel("Production buudget in $")
plt.ylabel("Global revenue in $")
plt.xlim(0, 450000000)
plt.ylim(0, 3000000000)
plt.show()
```

```python
regression = LinearRegression()
regression.fit(x,y)
```

Slope Coefficient

```python
regression.coef_  #theta_1
```

Intercept:

```python
regression.intercept_
```

```python
plt.figure(figsize = (20, 12))
plt.scatter(x,y, alpha=0.3)
plt.plot(x, regression.predict(x), color="red", linewidth=4)
#plt.plot(x,y) we are using regression.predict(x) , we are using the predicted x value instead of the real y value.


plt.title("Film production cost and global revenue")
plt.xlabel("Production buudget in $")
plt.ylabel("Global revenue in $")
plt.xlim(0, 450000000)
plt.ylim(0, 3000000000)
plt.show()
```
